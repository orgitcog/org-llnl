```python
# For D=0.01, dx=0.01, lambda=100: dt <= 1 / (2*0.01/0.01^2 + 100) = 1 / (20000 + 100) approx 5e-5

L = 1.0       # Length of the spatial domain [0, L]
T = 4.0       # Total simulation time.
N = 101       # Number of spatial points (includes boundaries). N-1 intervals.
D = 0.01      # Diffusion coefficient
dx = L / (N - 1)   # Spatial step size
print("dx", dx)
Lambda = 25.0 # Reaction rate constant (Note: large positive lambda leads to rapid growth).
M = 80000     # Number of time steps. M = T / dt --> dt = T/M
dt = T / M # Time step size

print("dt:", dt)

dt_RHS = 1 / ((2*D/(dx*dx)) + Lambda)
print("(2*D/(dx*dx))", (2*D)/(dx*dx))
print("dt_RHS", dt_RHS)
1/225
4/80000
```
