```python
output_file = "hdpy-bind-pdbbind-new.h5"
root_data_dir = "/g/g13/jones289/workspace/hd-cuda-master/datasets/pdbbind"
deepchem_aa_feat_path = Path(root_data_dir + "/deepchem_features/deepchem_baseline_feats.pkl")

train_3dcnn_feat_path = Path(root_data_dir + "/fusion_ml/3dcnn/train_3dcnn.npy")
test_3dcnn_feat_path = Path(root_data_dir + "/fusion_ml/3dcnn/test_3dcnn.npy")
train_sgcnn_feat_path = Path(root_data_dir + "/fusion_ml/sgcnn/train_sgcnn.npy")
test_sgcnn_feat_path = Path(root_data_dir + "/fusion_ml/sgcnn/test_sgcnn.npy")
train_fusion_feat_path = Path(root_data_dir + "/fusion_ml/fusion/train_fusion.npy")
test_fusion_feat_path = Path(root_data_dir + "/fusion_ml/fusion/test_fusion.npy")


# fusion_ml_sgcnn_feat_path = Path(root_data_dir + "/fusion_ml/sgcnn/sgcnn_data.npy")
# fusion_ml_fusion_feat_path = Path(root_data_dir + "/fusion_ml/fusion/fusion_data.npy")
pdbbind_2016_fps_feat_path = Path(root_data_dir + "/pdbbind_2016_fingerprints/pdbbind_2016_fps_new/v_2016_refined_pdbid_list.csv")


deepchem_aa_feat_dict = load_deepchem_feats(deepchem_aa_feat_path)
train_3dcnn_feat_dict = load_ml_fusion_feats_from_npy(train_3dcnn_feat_path,
                                                metadata_path="/g/g13/jones289/workspace/hd-cuda-master/datasets/pdbbind/fusion_ml/3dcnn/pdbbind_2016_train_metadata.csv")
test_3dcnn_feat_dict = load_ml_fusion_feats_from_npy(test_3dcnn_feat_path, 
                                                metadata_path="/g/g13/jones289/workspace/hd-cuda-master/datasets/pdbbind/fusion_ml/3dcnn/pdbbind_2016_test_metadata.csv")
train_fusion_feat_dict = load_ml_fusion_feats_from_npy(train_fusion_feat_path,
                                                metadata_path="/g/g13/jones289/workspace/hd-cuda-master/datasets/pdbbind/fusion_ml/3dcnn/pdbbind_2016_train_metadata.csv")
# import pdb
# pdb.set_trace()
test_fusion_feat_dict = load_ml_fusion_feats_from_npy(test_fusion_feat_path,
                                                metadata_path="/g/g13/jones289/workspace/hd-cuda-master/datasets/pdbbind/fusion_ml/3dcnn/pdbbind_2016_test_metadata.csv")


train_sgcnn_feat_dict = load_ml_fusion_feats_from_npy(train_sgcnn_feat_path,
                                                          metadata_path="/g/g13/jones289/workspace/hd-cuda-master/datasets/pdbbind/fusion_ml/3dcnn/pdbbind_2016_train_metadata.csv")
test_sgcnn_feat_dict = load_ml_fusion_feats_from_npy(test_sgcnn_feat_path,
                                                          metadata_path="/g/g13/jones289/workspace/hd-cuda-master/datasets/pdbbind/fusion_ml/3dcnn/pdbbind_2016_test_metadata.csv")

# fusion_ml_fusion_feat_dict = load_ml_fusion_feats(fusion_ml_fusion_feat_path)
fp_feat_dict = load_fp_feats(pdbbind_2016_fps_feat_path)
```

```python
def write_feats(dataset_name, data_dict):
    with h5py.File(output_file, 'a') as f:
        for pdbid in tqdm(data_dict, 
                          desc=f"writing {dataset_name} feats.."):
            
            if pdbid not in affinity_map.keys():
                continue
                
            pdbid_group = f.require_group(pdbid)
            pdbid_group.attrs['affinity'] = affinity_map[pdbid]
        
            data = data_dict[pdbid]
            pdbid_group.require_dataset(dataset_name,
                                   data=data,
                                    shape=data.shape,
                                    dtype=np.float32
                                   )
```

```python
def main():
    write_feats("deepchem-aa-feat", deepchem_aa_feat_dict)
    write_feats("fusion-3dcnn-feat", train_3dcnn_feat_dict)
    write_feats("fusion-3dcnn-feat", test_3dcnn_feat_dict)
    write_feats("fusion-feat", train_fusion_feat_dict)
    write_feats("fusion-feat", test_fusion_feat_dict)
    write_feats("fusion-sgcnn-feat", train_sgcnn_feat_dict)
    write_feats("fusion-sgcnn-feat", test_sgcnn_feat_dict)
    # write_feats("fusion-sgcnn-feat", fusion_ml_sgcnn_feat_dict)
    # write_feats("fusion-feat", fusion_ml_fusion_feat_dict)
    write_feats("ecfp-fingerprints",fp_feat_dict)
```

```python
main()
```
